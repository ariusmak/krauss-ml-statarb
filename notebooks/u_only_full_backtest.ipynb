# U-only Full Backtest (k=10)

Datastream US-only Phase 2 backtests for the U-only score family. This notebook runs the full k=10 backtest for each model family and reports:

- pre- and post-cost daily returns
- pre- and post-cost annualized returns
- pre- and post-cost Sharpe ratios
- per-period summary statistics


```python
import sys
from pathlib import Path

import pandas as pd
from IPython.display import display

sys.path.insert(0, str((Path('..') / 'src').resolve()))

from krauss.evaluation.phase2_ds_backtest_utils import (
    FAMILY_LABELS,
    FAMILIES,
    K,
    load_phase2_ds_data,
    per_period_stats,
    run_score_backtest,
    summary_stats,
)

pred, returns = load_phase2_ds_data()
pred.head()
```

```python
summary_rows = []
period_tables = []

for family in FAMILIES:
    _, _, daily = run_score_backtest(pred, returns, score_col=f'u_{family}', k=K)
    stats = summary_stats(daily)
    summary_rows.append({'Model': FAMILY_LABELS[family], **stats})

    period_df = per_period_stats(daily)
    period_df.insert(0, 'Model', FAMILY_LABELS[family])
    period_tables.append(period_df)

summary = pd.DataFrame(summary_rows).set_index('Model')
summary[['Daily Pre', 'Daily Post', 'Ann Pre', 'Ann Post']] = summary[['Daily Pre', 'Daily Post', 'Ann Pre', 'Ann Post']] * 100
display(summary.style.format({
    'Daily Pre': '{:.4f}%',
    'Daily Post': '{:.4f}%',
    'Ann Pre': '{:.2f}%',
    'Ann Post': '{:.2f}%',
    'Sharpe Pre': '{:.3f}',
    'Sharpe Post': '{:.3f}',
    'Days': '{:.0f}',
}).set_caption('U-only k=10 full-sample summary'))

per_period = pd.concat(period_tables, ignore_index=True)
display(per_period.style.format({
    'daily_pre': '{:.5f}',
    'daily_post': '{:.5f}',
    'ann_pre': '{:.3f}',
    'ann_post': '{:.3f}',
    'sharpe_pre': '{:.3f}',
    'sharpe_post': '{:.3f}',
}).set_caption('U-only k=10 per-period statistics'))
```
